```python
import os
import pandas as pd
import numpy as np
import kagglehub
import requests
from pathlib import Path
```

```python
# Download latest version
path = kagglehub.dataset_download("pratt3000/nisqa-corpus")

print("Path to dataset files:", path)
```

```python
for item in sorted(os.listdir(path)):
    print(item)
```

```python
mos_df = pd.read_csv(f"{path}/{item}/NISQA_TRAIN_LIVE/NISQA_TRAIN_LIVE_file.csv")
train_live_df = mos_df[mos_df["db"].str.contains("TRAIN_LIVE")]

print(f"Total TRAIN_LIVE files: {len(train_live_df)}")
print(f"\nColumns: {train_live_df.columns.tolist()}")
print(f"\nKey MOS dimensions:")
print(train_live_df[["mos", "noi", "col", "dis", "loud"]].describe())
```
